```python
import matplotlib.pyplot as plt
import numpy as np

buffersize = [10, 50, 100]
latency = [0.012, 0.08, 0.15]
debit = [83, 111, 125]

# log scale for latency and debit
latency = [np.log10(x) for x in latency]
debit = [np.log10(x) for x in debit]

# plot latency and debit as a function of buffer size
plt.plot(buffersize, latency, label='latency')
plt.plot(buffersize, debit, label='debit')
plt.xlabel('buffer size')
plt.ylabel('log(latency [ms/action]) / log(debit [actions/s])')
plt.legend()
```
